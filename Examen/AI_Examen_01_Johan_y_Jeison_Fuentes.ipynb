```python
# Apagar warnings en importacion de datos
import warnings
warnings.simplefilter('ignore', FutureWarning)
# Importa la libreria pandas
from pandas import *
import re
```

```python
data = read_csv('xmart.csv')
data
```

```python
limpiar =  data[(data['Year']<2019)].index
data.drop(limpiar, inplace=True)
data
```

```python
data.set_index('Country', inplace=True)
data.rename(columns={'Number of deaths due to tuberculosis, excluding HIV':'Cant. Muertes total',
					 'Deaths due to tuberculosis among HIV-negative people (per 100 000 population)':'Cant. Muertes 100mil'}, inplace=True)
data
```

```python
data['Cant. Muertes total'] = data['Cant. Muertes total'].str.extract(r"(^\d+)")
data['Cant. Muertes total'] = to_numeric(data['Cant. Muertes total'])
data
```

```python
data['Cant. Muertes 100mil'] = data['Cant. Muertes 100mil'].str.extract(r"(.*\d\s)")
data['Cant. Muertes 100mil'] = to_numeric(data['Cant. Muertes 100mil'])
data
```

```python
population = round(data['Cant. Muertes total'] * 100 / data['Cant. Muertes 100mil'])
data['Poblacion (1000s)'] = population
data
```

```python
data = data.fillna(value=0)
data.sort_values('Cant. Muertes total')
```
